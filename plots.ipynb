```python
import torch
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
problem = "ChainPendulum1.np"
files = {fname.split('_')[2]:fname for fname in glob.glob('biases/rollout_errs_*'+problem)}
```

```python
files
```

```python
for model, fname in files.items():
    rel_errs = pickle.load(open(fname,'rb')).cpu().data.numpy()
    plt.plot(rel_errs.mean(0)[...,0],label=model)
    plt.yscale('log')
#plt.plot(rel_errs.mean(0)[...,1],label='GT+eps')
plt.axvline(5)
plt.ylim((1e-7,1))
plt.legend()
```
